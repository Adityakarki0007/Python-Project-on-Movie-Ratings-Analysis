```python
import pandas as pd
```

```python
movie = pd.read_csv("C:\\Users\\ADI\\Downloads\\Python CSV\\R-Programming-A-Z-master\\Movie-Ratings.csv")
```

```python
movie.columns = ['Film','Genre','CriticRating','AudienceRating','BudgetMillions','Year'] #Changed the name of Headers
```

```python
movie
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
#lm_plot

sns.set_style('darkgrid') #changing the backgroubd of graph

lm = sns.lmplot(data=movie , x='CriticRating' , y='AudienceRating',fit_reg=False, hue='Genre',  aspect=1 ,scatter_kws={"s":50})
```

```python
#dist_plot
dist = sns.distplot(movie.AudienceRating)
```

```python
#kde_plot

sns.set_style('white')
k1 = sns.kdeplot(data=movie, x="CriticRating" , y="AudienceRating" , shade=True,shade_lowest=False,cmap='Reds')
```

```python
k2 = sns.kdeplot(data=movie, x="BudgetMillions" , y="AudienceRating" , shade=True,shade_lowest=False,cmap='Blues')
```

```python
k3 = sns.kdeplot(data=movie, x="BudgetMillions" , y="CriticRating" , shade=True,shade_lowest=False,cmap='Oranges')
```

```python
#Subplot 

f , axes =plt.subplots( 1, 2, figsize=(12,6) , sharex=True,sharey=True)
k1 = sns.kdeplot(data=movie, x="BudgetMillions" , y="AudienceRating" , shade=True,shade_lowest=False,cmap='Blues' , ax=axes[0])
k2 = sns.kdeplot(data=movie, x="BudgetMillions" , y="CriticRating" , shade=True,shade_lowest=False,cmap='Oranges', ax=axes[1])
k1.set(xlim=(-20,160))
```

```python
#violin_plot
v = sns.violinplot(data=movie, x='Genre' , y='CriticRating')
```

```python
#box_plot
box = sns.boxplot(data=movie,x='Genre',y='CriticRating')
```

```python
#FacetGrid

g = sns.FacetGrid(movie, row = 'Genre' , col = 'Year' , hue='Genre')
g = g.map(plt.scatter, 'CriticRating' , 'AudienceRating')
```

```python
g = sns.FacetGrid(movie, row = 'Genre' , col = 'Year' , hue='Genre')
g = g.map(plt.hist , 'BudgetMillions')
```

```python
#Putting Diagonals and coordinates
```

```python
g = sns.FacetGrid(movie, row = 'Genre' , col = 'Year' , hue='Genre')
g = g.map(plt.scatter, 'CriticRating' , 'AudienceRating')
g.set( xlim=(0,100), ylim=(0,100) )
for ax in g.axes.flat:
    ax.plot( (0,100) , (0,100) , c='gray', ls='--')
g.add_legend()
```
