```python
list1= list() #will use this as a Loop value
mylabels = list() #will use this for legend
for  gen in movie.Genre.unique():
    list1.append(movie[movie.Genre == gen].BudgetMillions)
    mylabels.append(gen)
    
h = plt.hist(list1,bins=30,stacked=True,rwidth=1,label=mylabels) #now all unique value of (Genre) are on loop
plt.legend()
plt.show()
```
